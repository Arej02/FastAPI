```python
import numpy as np
import pandas as pd
```

```python
adult=pd.read_csv("adult.csv",na_values="?")
adult.shape
```

```python
adult.sample(4)
```

```python
adult.drop(columns=['fnlwgt','education'],inplace=True)
```

```python
adult['native-country'].value_counts()
```

```python
adult['native-country'] = adult.apply(lambda x: 'US' if x['native-country'] == 'US' else 'Non-US Country', axis=1)
```

```python
adult.isnull().mean()*100
```

```python
adult.dropna(subset=['workclass','occupation'],inplace=True)
```

```python
adult.duplicated().sum()
```

```python
adult.drop_duplicates(inplace=True)
```

```python
adult.sample(4)
```

```python
adult['workclass'].unique()
```

```python
def workclass(row):
    if row in ['Local-gov','State-gov','Federal-gov']:
        return 'Government'
    elif row=='Private':
        return 'Private'
    elif row in ['Without-pay','Self-emp-not-inc','Self-emp-inc']:
        return 'Unemployed'
    else:
        return 'Other'
```

```python
adult['workclass']=adult['workclass'].apply(workclass)
```

```python
adult['marital-status'].unique()
```

```python
def marital_status(row):
    if row in ['Separated','Divorced','Married-spouse-absent']:
        return 'Divorced'
    elif row=='Never-married':
        return 'Single'
    elif row in ['Married-civ-spouse','Married-AF-spouse']:
        return 'Married'
    elif row=='Widowed':
        return "Widow"
    else:
        return 'Other'
```

```python
adult['marital-status']=adult['marital-status'].apply(marital_status)
```

```python
adult['occupation'].unique()
```

```python
def job_category(row):
    if row in ['Exec-managerial']:
        return 'Management'
    elif row in ['Prof-specialty', 'Tech-support']:
        return 'Professional'
    elif row in ['Other-service', 'Sales', 'Adm-clerical', 'Priv-house-serv']:
        return 'Services'
    elif row in ['Farming-fishing', 'Craft-repair', 'Handlers-cleaners', 'Transport-moving', 'Machine-op-inspct']:
        return 'Manual Labor'
    elif row == 'Protective-serv':
        return 'Protective Services'
    elif row == 'Armed-Forces':
        return 'Armed Forces'
    else:
        return 'Other'
```

```python
adult['occupation']=adult['occupation'].apply(job_category)
```

```python
adult.head()
```

```python
X=adult.drop(columns=['income'])
Y=adult['income']
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=10,test_size=0.3)
```

```python
numerical_features=[val for val in adult.columns if adult[val].dtype in ['int64']]
numerical_features
```

```python
exclude=['income','workclass']
categorical_features=[val for val in adult.columns if adult[val].dtype=='object' and val not in exclude]
categorical_features
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder,LabelEncoder,StandardScaler

transformer=ColumnTransformer(transformers=[
    ('trf1',OrdinalEncoder(categories=[['Government','Private','Unemployed']]),['workclass']),
    ('trf2',OneHotEncoder(drop='first',sparse_output=False),categorical_features),
    ('trf3',StandardScaler(),numerical_features)],remainder='passthrough')
```

```python
label=LabelEncoder()

y_train_encoded=label.fit_transform(Y_train)
y_test_encoded=label.transform(Y_test)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier()
pipe=make_pipeline(transformer,rfc)
```

```python
pipe.fit(X_train,y_train_encoded)
```

```python
from sklearn.metrics import f1_score

y_pred=pipe.predict(X_test)
print(f"The f1 score is:{f1_score(y_pred,y_test_encoded)}")
```

```python
import pickle
with open("model.pkl", "wb") as f:
    pickle.dump(pipe, f)
with open("label_encoder.pkl", "wb") as f:
    pickle.dump(label, f)
```
